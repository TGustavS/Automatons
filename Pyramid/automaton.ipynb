```python
import random
def convert_rule(rule):
    rule = bin(rule)[2:]
    rule = (8-len(rule)) * '0' + rule
    rule = rule.replace('0',' ')
    return rule.replace('1','#')

def update(tem,rule):
    current = [' ' for i in range(len(tem))]
    tem.append(tem[-1])
    for i in range(len(tem)-1):
        #111
        if tem[i-1] == '#' and tem[i] == '#' and tem[i+1] == '#':
            current[i] = rule[0]
        #110
        elif tem[i-1] == '#' and tem[i] == '#' and tem[i+1] == ' ':
            current[i] = rule[1]
        #101
        elif tem[i-1] == '#' and tem[i] == ' ' and tem[i+1] == '#':
            current[i] = rule[2]
        #100
        elif tem[i-1] == '#' and tem[i] == ' ' and tem[i+1] == ' ':
            current[i] = rule[3]
        #011
        elif tem[i-1] == ' ' and tem[i] == '#' and tem[i+1] == '#':
            current[i] = rule[4]
        #010
        elif tem[i-1] == ' ' and tem[i] == '#' and tem[i+1] == ' ':
            current[i] = rule[5]
        #001
        elif tem[i-1] == ' ' and tem[i] == ' ' and tem[i+1] == '#':
            current[i] = rule[6]
        #000
        elif tem[i-1] == ' ' and tem[i] == ' ' and tem[i+1] == ' ':
            current[i] = rule[7]
    return current

def automaton(rule,n,rand = 0):
    rule = convert_rule(rule)
    first = [' ' for i in range(2*n+1)]
    if rand == 1:
        for i in range(len(first)):
            r = random.randint(0,1)
            if r == 1:
                first[i] = '#'
    elif rand == 2:
        first[n-30] = '#'
        first[n+30] = '#'
    else:
        first[n] = '#'    
    tem = first
    for i in range(n):
        if len(tem) < 114:
            for l in range(len(tem)):
                if tem[l] == '#':
                    print(chr(9608),end='')
                else:
                    print(' ',end='')
        else:
            for l in range(int(len(tem)/2)-57,int(len(tem)/2)+57):
                if tem[l] == '#':
                    print(chr(9608),end='')
                else:
                    print(' ',end='')
        tem = update(tem,rule)
        print()
automaton(30,1000)
```

```python
for i in range(255):
    print('Rule',i)
    print(automaton(i,30))
```
